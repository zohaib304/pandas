```python
import pandas as pd # pandas is a data analysis library
```

```python
data = pd.read_csv('data/survey_results_public.csv', index_col='Respondent') # Read in the data
schema = pd.read_csv('data/survey_results_schema.csv', index_col='Column') # Read in the schema
```

```python
data.columns # to see the column names
```

```python
data.columns = [x.lower() for x in data.columns] # to make all the column names lowercase
```

```python
data.columns = [x.capitalize() for x in data.columns] # to capitalize the first letter of each column name 
```

```python
data.columns
```

## Rename Column `ConvertedComp` to `SalaryUSD`

```python
data.rename(columns={"Convertedcomp": 'SalaryUSD'}, inplace=True) # to rename a column
```

```python
data.columns
```

```python
data['SalaryUSD']
```

```python
data['Hobbyist']
```

```python
data['Hobbyist'] = data['Hobbyist'].map({'Yes': True, 'No': False}) # to convert a column to boolean
```

```python
data
```

# Sorting data by country

```python
# to sort the data by a column
data.sort_values(by=['Country', 'SalaryUSD'], ascending=[True, False], inplace=True)
```

```python
data[['Country', 'SalaryUSD']].head(10)
```

```python
data['SalaryUSD'].nlargest(10)
```

```python
data.nlargest(10, 'SalaryUSD')
```

```python
filt = data['Age'] > 20 # filter out the rows where the age is greater than 20
data.loc[filt, 'Age']
```
