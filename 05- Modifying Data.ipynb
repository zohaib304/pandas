# Examples

```python
people = {
    "first" : ["Alfred", "Alice", "Bob", "Carol"],
    "last" : ["Brown", "Smith", "Jones", "Williams"],
    "age" : [17, 13, 44, 42],
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns
```

If we want to update the column name of the series, we can use the following command:

```python
df.columns = ['first_name', 'last_name', 'age']
```

```python
df
```

List comprehnsion can be use to update the column name of the series. Like remove space or make upper and lower case.

```python
df.columns = [x.lower() for x in df.columns]
df
```

To rename the column we can use df.rename(columns={'column_name': 'new_column_name'}) and pass the dictionary.

```python
df.rename(columns={'first_name' : 'first', 'last_name' : 'last'}, inplace=True)
```

```python
df
```

Now lets change the last name is second row.

```python
df.loc[2] = ['Bob', 'Tariq', 23]
df
```

```python
df.loc[2, ['last', 'age']] = ['Jones', 44]
df
```

```python
df.loc[2, 'last'] = 'Mike'
df
```

```python
df.at[2, 'last'] = 'Jones'
df
```

```python
filt = (df['age'] == 42)
df[filt]['last'] = 'smith'
```

```python
filt = (df['age'] == 42) #
df.loc[filt, 'last'] = 'Tait' # to change the value of the last name
df
```

```python
df['last'] = df['last'].str.lower() # to convert all the last names to lowercase
df
```

There four method 
-   apply()
-   applymap()
-   map()
-   replace()

Appy is used for calling a function on each element of the series/dataframe.

```python
df['last'].apply(len) # to get the length of the last names
```

```python
def update_last(last):
    return last.upper()
```

```python
df['last'].apply(update_last) # to convert all the last names to uppercase
```

```python
df['last'] = df['last'].apply(update_last) # to convert all the last names to uppercase
df
```

```python
df['first'].replace({'Alfred' : 'Chris', 'Bob' : 'Mary'}) # to map the first names
```
