# Grouping 

Split the Object -> Apply a function -> combile those results

```python
country_grp = data.groupby(['Country'])
```

```python
country_grp.get_group('Pakistan')
```

```python
filt = data['Country'] == 'Pakistan'
data.loc[filt]['SocialMedia'].value_counts()
```

```python
country_grp['SocialMedia'].value_counts(normalize=True).loc['China']
```

```python
filt = data['Country'] == 'Pakistan'
data.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_grp['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondets = data['Country'].value_counts()
country_respondets
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(
    lambda x: x.str.contains('Python').sum())

country_uses_python
```

```python
python_df = pd.concat((country_respondets, country_uses_python), axis='columns', sort=False) # concatenate the two dataframes
python_df.head()
```

```python
    python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True) # rename the columns
    python_df.head()
```

# calculate the percentage

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython'] / python_df['NumRespondents']) * 100 # calculate the percentage
```

```python
python_df.head() # to see the first 5 rows
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True) # sort the dataframe
```
